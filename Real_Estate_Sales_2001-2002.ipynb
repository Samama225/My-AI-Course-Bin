```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"D:\Car\Customer_Segmentation\Real_Estate_Sales_2001-2022_GL-Short.csv")
```

```python
df.head()
```

```python
print("Mean: ",np.mean(df["Sale Amount"]))
print("Average: ",np.average(df["Sale Amount"]))
print("Standard Deviation: ",np.std(df["Sale Amount"]))
print("Median: ",np.median(df["Sale Amount"]))
print("25th percentile: ",np.percentile(df["Sale Amount"],25))
print("75th percentile: ",np.percentile(df["Sale Amount"],75))
print("3rd percentile: ",np.percentile(df["Sale Amount"],3))
print("min: ",np.min(df["Sale Amount"]))
print("max: ",np.max(df["Sale Amount"]))
print("Square: ",np.square(df["Sale Amount"]))
print("Square Root: ",np.sqrt(df["Sale Amount"]))
print("Power: ",np.power(df["Sale Amount"],2))
print("Absolute: ",np.abs(df["Sale Amount"]))
```

```python
Unspecified=df["Non Use Code"].isnull()
df.loc[Unspecified,"Non Use Code"]=np.random.choice(df["Non Use Code"].dropna(),size=Unspecified.sum())

Unspecified=df["bath"].isnull()
df.loc[Unspecified,"bath"]=np.random.choice(df["bath"].dropna(),size=Unspecified.sum())

Unspecified=df["acre_lot"].isnull()
df.loc[Unspecified,"acre_lot"]=np.random.choice(df["acre_lot"].dropna(),size=Unspecified.sum())

Unspecified=df["house_size"].isnull()
df.loc[Unspecified,"house_size"]=np.random.choice(df["house_size"].dropna(),size=Unspecified.sum())

df.drop("prev_sold_date",axis=1,inplace=True)
```

```python
addition=df["Sale Amount"]+df["Sale Amount"]
print("Addition: ",addition)
subtraction=df["Sale Amount"]-df["Sale Amount"]
print("Subtraction: ",subtraction)
multiplication=df["Sale Amount"]*df["Sale Amount"]
print("Multiplication: ",multiplication)
division=df["Sale Amount"]/df["Sale Amount"]
print("Division",division)
modulos=df["Sale Amount"]%df["Sale Amount"]
print("Modulos: ",modulos)
```

```python
pricePie = (df["Sale Amount"]/np.pi) +1
sine_values = np.sin(pricePie)
print("sine: ",sine_values)
cosine_values = np.cos(pricePie)
print("cosine: ",cosine_values)
tangent_values = np.tan(pricePie)
print("tangent: ",tangent_values)
print("Exponential: ",np.exp(df["Sale Amount"]))
```

```python
log_array = np.log(pricePie)
print("log_array: ",log_array)
log10_array = np.log10(pricePie)
print("log10_array: ",log10_array)
sinh_values = np.sinh(pricePie)
print("Sine h: ", sinh_values)
cosh_values = np.cosh(pricePie)
print("Cosine h: ", cosh_values)
tanh_values = np.tanh(pricePie)
print("Tangent h: ", tanh_values)
asinh_values = np.arcsinh(pricePie)
print("ASine h:", asinh_values)
acosh_values = np.arccosh(pricePie)
print("ACosine h:", acosh_values)
```

```python
sns.countplot(data=df,y="TOWN",order=df["TOWN"].value_counts().index)
plt.title("Number of Property Sales by Town")
plt.xlabel("Count")
plt.ylabel("Town")
plt.show()
```

```python
sns.histplot(df["Sale Amount"],bins=30,kde=True)
plt.title("Distribution of Sale Amount")
plt.xlabel("Sale Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.countplot(data=df,x="Property_Type")
plt.title("Property Type Distribution")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(data=df.reset_index(drop=True), x="Residential_Type")
plt.title("Residential Type Count")
plt.show()
```

```python
sns.boxplot(data=df,x="Property_Type",y="Sale Amount")
plt.title("Sale Amount by Property Type")
plt.xticks(rotation=45)
plt.show()
```

```python
top_towns = (df.groupby("TOWN")["Sale Amount"].mean().sort_values(ascending=False).head(10).reset_index())
plt.figure(figsize=(10,5))
sns.barplot(data=top_towns,x="TOWN",y="Sale Amount")
plt.title("Top 10 Towns by Average Sale Amount")
plt.xticks(rotation=45)
plt.show()
```

```python
numeric_df = df.select_dtypes(include=["int64", "float64"])
sns.heatmap(numeric_df.corr(),annot=True,cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
df["Date Recorded"] = pd.to_datetime(df["Date Recorded"])
df["Year"] = df["Date Recorded"].dt.year
yearly_sales = df["Year"].value_counts().sort_index()
plt.figure(figsize=(12,5))
sns.lineplot(x=yearly_sales.index,y=yearly_sales.values,marker="o")
plt.title("Yearly Property Sales Trend")
plt.xlabel("Year")
plt.ylabel("Number of Sales")
plt.show()
```
