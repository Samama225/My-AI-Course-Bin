```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
driver = webdriver.Chrome()
```

```python
names = []
prices = []
links = []
sold_counts = []
```

```python
for page in range(1, 6):
    url = f"https://www.daraz.pk/catalog/?q=Smart+Phones&page={page}"
    print("Scraping Page:", page)
    driver.get(url)
   
    time.sleep(5)

    product_elements = driver.find_elements(By.XPATH,'//div[@data-qa-locator="product-item"]')
    print("Products Found:", len(product_elements))

    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")

    products = soup.find_all("div",attrs={"data-qa-locator": "product-item"})

    for product in products:
        try:
            name = product.find("div", class_="RfADt").get_text(strip=True)
        except:
            name = "N/A"

        try:
            price = product.find("span", class_="ooOxS").get_text(strip=True)
        except:
            price = "N/A"

        try:
            link = product.find("a")["href"]

            if not link.startswith("http"):
                link = "https:" + link
        except:
            link = "N/A"

        try:
            sold = product.find("span", class_="_1cEkb").get_text(strip=True)
        except:
            sold = "N/A"

        names.append(name)
        prices.append(price)
        links.append(link)
        sold_counts.append(sold)
```

```python
df = pd.DataFrame({"Product Name": names,"Price": prices,"Sold Items": sold_counts,"Product Link": links})
df.to_csv("daraz_smartphones.csv", index=False)
print("CSV File Saved Successfully")
```
