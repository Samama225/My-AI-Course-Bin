```python
df["postalCode"]=pd.to_numeric(df["postalCode"],errors="coerce")
```

```python
print("Mean: ",np.mean(df["postalCode"]))
print("Average: ",np.average(df["postalCode"]))
print("Standard Deviation: ",np.std(df["postalCode"]))
print("Median: ",np.median(df["postalCode"]))
print("25th percentile: ",np.percentile(df["postalCode"],25))
print("75th percentile: ",np.percentile(df["postalCode"],75))
print("3rd percentile: ",np.percentile(df["postalCode"],3))
print("min: ",np.min(df["postalCode"]))
print("max: ",np.max(df["postalCode"]))
print("Square: ",np.square(df["postalCode"]))
print("Square Root: ",np.sqrt(df["postalCode"]))
print("Power: ",np.power(df["postalCode"],2))
print("Absolute: ",np.abs(df["postalCode"]))
```

```python
addition=df["postalCode"]+df["postalCode"]
print("Addition: ",addition)
subtraction=df["postalCode"]-df["postalCode"]
print("Subtraction: ",subtraction)
multiplication=df["postalCode"]*df["postalCode"]
print("Multiplication: ",multiplication)
division=df["postalCode"]/df["postalCode"]
print("Division",division)
modulos=df["postalCode"]%df["postalCode"]
print("Modulos: ",modulos)
```

```python
pricePie = (df["postalCode"]/np.pi) +1
sine_values = np.sin(pricePie)
print("sine: ",sine_values)
cosine_values = np.cos(pricePie)
print("cosine: ",cosine_values)
tangent_values = np.tan(pricePie)
print("tangent: ",tangent_values)
print("Exponential: ",np.exp(df["postalCode"]))
```

```python
log_array = np.log(pricePie)
print("log_array: ",log_array)
log10_array = np.log10(pricePie)
print("log10_array: ",log10_array)
sinh_values = np.sinh(pricePie)
print("Sine h: ", sinh_values)
cosh_values = np.cosh(pricePie)
print("Cosine h: ", cosh_values)
tanh_values = np.tanh(pricePie)
print("Tangent h: ", tanh_values)
asinh_values = np.arcsinh(pricePie)
print("ASine h:", asinh_values)
acosh_values = np.arccosh(pricePie)
print("ACosine h:", acosh_values)
```
