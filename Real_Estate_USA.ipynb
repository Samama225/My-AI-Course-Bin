```python
log_array = np.log(pricePie)
print("log_array: ",log_array)
log10_array = np.log10(pricePie)
print("log10_array: ",log10_array)
sinh_values = np.sinh(pricePie)
print("Sine h: ", sinh_values)
cosh_values = np.cosh(pricePie)
print("Cosine h: ", cosh_values)
tanh_values = np.tanh(pricePie)
print("Tangent h: ", tanh_values)
asinh_values = np.arcsinh(pricePie)
print("ASine h:", asinh_values)
acosh_values = np.arccosh(pricePie)
print("ACosine h:", acosh_values)
```

```python
Unspecified=df["bed"].isnull()
df.loc[Unspecified,"bed"]=np.random.choice(df["bed"].dropna(),size=Unspecified.sum())

Unspecified=df["bath"].isnull()
df.loc[Unspecified,"bath"]=np.random.choice(df["bath"].dropna(),size=Unspecified.sum())

Unspecified=df["acre_lot"].isnull()
df.loc[Unspecified,"acre_lot"]=np.random.choice(df["acre_lot"].dropna(),size=Unspecified.sum())

Unspecified=df["house_size"].isnull()
df.loc[Unspecified,"house_size"]=np.random.choice(df["house_size"].dropna(),size=Unspecified.sum())

df.drop("prev_sold_date",axis=1,inplace=True)
```

```python
Q1=df["price"].quantile(0.25)
Q3=df["price"].quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR

df=df[(df["price"] >= lower_bound) & (df["price"] <= upper_bound)]
```

```python
num_cols = ['price', 'Bed', 'Bath', 'ACRE_lot', 'house_size']

for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(
    df[num_cols].corr(),
    annot=True,
    fmt=".2f"
)
plt.title("Correlation Heatmap")
plt.show()
```

```python
for col in num_cols:
    plt.figure(figsize=(6,3))
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(
    x='house_size',
    y='price',
    data=df
)
plt.title("Price vs House Size")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.barplot(
    x='Bed',
    y='price',
    data=df,
    estimator='mean'
)
plt.title("Average Price by Bedrooms")
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.boxplot(
    x='state',
    y='price',
    data=df
)
plt.xticks(rotation=90)
plt.title("Price Distribution Across States")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(
    y=df['city'],
    order=df['city'].value_counts().head(10).index
)
plt.title("Top 10 Cities")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.regplot(
    x='ACRE_lot',
    y='price',
    data=df
)
plt.title("Acre Lot vs Price")
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.violinplot(
    x='state',
    y='house_size',
    data=df
)
plt.xticks(rotation=90)
plt.title("House Size Distribution by State")
plt.show()
```
