```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
driver = webdriver.Chrome()
```

```python
names = []
prices = []
ratings = []
reviews = []
links = []
```

```python
for page in range(1, 6):
    url = f"https://www.amazon.com/s?k=smart+home+devices&page={page}"
    print("Opening Page:", page)
    driver.get(url)
    
    time.sleep(5)

    products = driver.find_elements(By.XPATH,'//div[@data-component-type="s-search-result"]')
    print("Products Found:", len(products))
    soup = BeautifulSoup(driver.page_source, "html.parser")
    items = soup.find_all("div",attrs={"data-component-type": "s-search-result"})

    for item in items:

        try:
            name = item.find("h2").get_text(strip=True)
        except:
            name = "N/A"

        try:
            link = item.find("h2").find("a")["href"]
            link = "https://www.amazon.com" + link
        except:
            link = "N/A"

        try:
            whole = item.find("span", class_="a-price-whole").text
            fraction = item.find("span", class_="a-price-fraction").text
            price = whole + "." + fraction
        except:
            price = "N/A"

        try:
            rating = item.find("span", class_="a-icon-alt").text
        except:
            rating = "N/A"

        try:
            review = item.find("span", class_="a-size-base s-underline-text").text
        except:
            review = "N/A"

        names.append(name)
        prices.append(price)
        ratings.append(rating)
        reviews.append(review)
        links.append(link)
```

```python
df = pd.DataFrame({"Product Name": names,"Price": prices,"Rating": ratings,"Reviews": reviews,"Product Link": links})
df.to_csv("amazon_smart_home_devices.csv", index=False)
```
