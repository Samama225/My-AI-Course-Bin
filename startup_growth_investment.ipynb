```python
print("Mean: ",np.mean(df["Number of Investors"]))
print("Average: ",np.average(df["Number of Investors"]))
print("Standard Deviation: ",np.std(df["Number of Investors"]))
print("Median: ",np.median(df["Number of Investors"]))
print("25th percentile: ",np.percentile(df["Number of Investors"],25))
print("75th percentile: ",np.percentile(df["Number of Investors"],75))
print("3rd percentile: ",np.percentile(df["Number of Investors"],3))
print("min: ",np.min(df["Number of Investors"]))
print("max: ",np.max(df["Number of Investors"]))
print("Square: ",np.square(df["Number of Investors"]))
print("Square Root: ",np.sqrt(df["Number of Investors"]))
print("Power: ",np.power(df["Number of Investors"],2))
print("Absolute: ",np.abs(df["Number of Investors"]))
```

```python
addition=df["Number of Investors"]+df["Number of Investors"]
print("Addition: ",addition)
subtraction=df["Number of Investors"]-df["Number of Investors"]
print("Subtraction: ",subtraction)
multiplication=df["Number of Investors"]*df["Number of Investors"]
print("Multiplication: ",multiplication)
division=df["Number of Investors"]/df["Number of Investors"]
print("Division",division)
modulos=df["Number of Investors"]%df["Number of Investors"]
print("Modulos: ",modulos)
```

```python
pricePie = (df["Number of Investors"]/np.pi) +1
sine_values = np.sin(pricePie)
print("sine: ",sine_values)
cosine_values = np.cos(pricePie)
print("cosine: ",cosine_values)
tangent_values = np.tan(pricePie)
print("tangent: ",tangent_values)
print("Exponential: ",np.exp(df["Number of Investors"]))
```

```python
log_array = np.log(pricePie)
print("log_array: ",log_array)
log10_array = np.log10(pricePie)
print("log10_array: ",log10_array)
sinh_values = np.sinh(pricePie)
print("Sine h: ", sinh_values)
cosh_values = np.cosh(pricePie)
print("Cosine h: ", cosh_values)
tanh_values = np.tanh(pricePie)
print("Tangent h: ", tanh_values)
asinh_values = np.arcsinh(pricePie)
print("ASine h:", asinh_values)
acosh_values = np.arccosh(pricePie)
print("ACosine h:", acosh_values)
```

```python
sns.countplot(data=df, x="Industry")
plt.title("Number of Startups by Industry")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.scatterplot(data=df,x="Funding_Rounds",y="Valuation (USD)",hue="Industry")
plt.title("Funding Rounds vs Valuation")
plt.show()
```

```python
avg_growth = df.groupby("Country")["Growth Rate (%)"].mean().reset_index()

sns.barplot(data=avg_growth,x="Country",y="Growth Rate (%)")
plt.title("Average Growth Rate by Country")
plt.xticks(rotation=45)
plt.show()
```
