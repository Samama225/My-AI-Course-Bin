```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
driver = webdriver.Chrome()
```

```python
data = []

for page in range(1, 6):
    url = f"https://www.ebay.com/b/Cell-Phones-Smartphones/9355/bn_320094?_pgn={page}"
    print("Opening Page:", page)
    driver.get(url)
    time.sleep(8)

    soup = BeautifulSoup(driver.page_source, "html.parser")
    products = soup.find_all("div", class_="brwrvr__item-card__signals")
    print("Products Found:", len(products))
    for product in products:
        try:
            title = product.find("h3",class_="textual-display bsig__title__text").get_text(strip=True)
        except:
            title = "N/A"
        if title == "Shop on eBay":
            continue
        try:
            price = product.find("span",class_="textual-display bsig__price bsig__price--displayprice").get_text(strip=True)
        except:
            price = "N/A"
        try:
            shipping = product.find("span",class_="textual-display bsig__generic bsig__logisticsCost").get_text(strip=True)
        except:
            shipping = "N/A"
        try:
            link = product.find("a")["href"]
        except:
            link = "N/A"
        data.append([title, price, shipping, link])
```

```python
df = pd.DataFrame(data, columns=["Product Title","Price","Shipping","Product Link"])
df.drop_duplicates(inplace=True)
df.to_csv("ebay_smartphones.csv", index=False, encoding="utf-8")
```
